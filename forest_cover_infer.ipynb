```python
# Import necessary libraries

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
from sklearn.pipeline import Pipeline
```

```python
# Load dataset 

forest = pd.read_csv("Dataset location", index_col=0)
```

```python
forest.dtypes
```

```python
# Extracting 'Wilderness_Area' from one-hot encoded columns 

forest['Wilderness_Area'] = forest.iloc[:,10:14].values.argmax(axis=1) + 1
forest[['Wilderness_Area', 'Wilderness_Area1', 'Wilderness_Area2', 'Wilderness_Area3', 'Wilderness_Area4']]
```

```python
forest.iloc[:,-42:-2]
```

```python
# Extracting 'Soil_Type' from one-hot encoded columns

forest['Soil_Type'] = forest.iloc[:,-42:-2].values.argmax(axis=1) + 1
forest.iloc[:, -43:]
```

```python
forest.columns
```

```python
# Selecting relevant features

forest = forest[['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology',
       'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways',
       'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
       'Horizontal_Distance_To_Fire_Points','Cover_Type', 'Wilderness_Area',
       'Soil_Type' ]]
forest.dtypes
```

```python
# Splitting dataset into features (X) and target variable (y)

X, y = forest.drop('Cover_Type', axis=1), forest['Cover_Type']
```

```python
# Setting up OneHotEncoder for categorical variables

ohc = OneHotEncoder(sparse_output=False).set_output(transform='pandas')

# Column transformer to apply transformations to selected features
ct = make_column_transformer(('passthrough',['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology',
       'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways',
       'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
       'Horizontal_Distance_To_Fire_Points']),
       (ohc,['Wilderness_Area','Soil_Type']),
       verbose_feature_names_out=False).set_output(transform='pandas')

# Transforming features
enc_forest_X = ct.fit_transform(X)

```

```python
enc_forest_X.shape
```

```python
# Initializing and training Random Forest Classifier
rf = RandomForestClassifier(random_state=24, max_features=10)
rf.fit(enc_forest_X, y)
```

```python
# Creating a pipeline for transformation and prediction
pipe = Pipeline([('CT',ct),('RF',rf)])
pipe.fit(X,y)
```

```python
# Define path to save the trained model in Google Drive

import pickle
file_model = open("give you relevent location", "wb")
pickle.dump(pipe, file_model)
file_model.close()
```
